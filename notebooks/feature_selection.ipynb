```python
y_binary = pca_analysis.y_binary
x = pca_analysis.x
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd 
```

```python
feature_names = x.columns.tolist()
forest = RandomForestClassifier(n_estimators=100 , random_state=42)
forest.fit(x,y_binary)
importances = forest.feature_importances_
feature_importance_df = pd.DataFrame({'feature': feature_names , 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)
```

```python
feature_importance_df
```

```python
selected_features = feature_importance_df['feature'].head(9).tolist()
x_selected_RF = x[selected_features]
```

```python
x_selected_RF
```

```python
from sklearn.feature_selection import RFE , SelectKBest , chi2
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=100)
rfe = RFE(model,n_features_to_select=9)
fit = rfe.fit(x,y_binary)
selected_features_rfe = x.columns[fit.support_]
```
